# TWO SUM PROBLEM


Example 1:

Input: nums = [2,7,11,15], target = 9
Output: [0,1]
Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].


Example 2:

Input: nums = [3,2,4], target = 6
Output: [1,2]


Example 3:

Input: nums = [3,3], target = 6
Output: [0,1]


```python
# Brute Force Method - Try every possible option unitl you find the correct one.

nums = [5,9,1,2,4,15,6,3]        
target = 13

for i in range(0,len(nums)-1):
    for j in range(i + 1, len(nums)):
        if nums[i] + nums[j] == target:
            print(i,j)  # return [i, j]


# Time Complexity = O(n^2)
# Space Complexity = O(1)
```

```python
# Method 2 (Optimal)

dict = {}
for i in range (0, len(nums)):
    remaining = target - nums[i]
    if remaining in dict :
        print(dict[remaining], i)
    dict[nums[i]]= i
    
# Time Complexity = O(n)
# Space Complexity = O(1)
```

# Armstrong Number

153 - [1^3 + 5^3 + 3^3] -> [1 + 125 + 27] = 153 <br>
1634 - [1^4 + 6^4 + 3^4 + 4^4] -> 1634

```python
n = 153
num = n
num2 = n
count = 0
ans = 0

while num > 0:
    last_digit = num % 10
    num = num // 10
    count += 1

while num2 > 0:
    last_digit = num2 % 10
    num2 = num2 // 10
    ans = ans + (last_digit ** count)
print(ans)

# TC = O(log n)
# SC = O(1)
```

```python
# Method 2 

n = 153
num = n
ans = 0
count = len(str(num))

while num > 0:
    last_digit = num % 10
    num = num // 10
    ans = ans + (last_digit ** count)
print(ans)
```

# Print Factors / Divisors
10 - [1,2,5,10] <br>
15 - [1,3,5,15] <br>
25 - [1,5,25] <br>
7 - [1,7] <br>
19 - [1,19] <br>



```python
n = 10
lst = []
for i in range(1,n+1):
    if n % i == 0:
        lst.append(i)
print(lst)

# TC = O(n)
# SC = O(K) - k would be the total number of factors e.g. [1,2,5,10]
```

```python
# Method 2 

num = 10
result = []

for i in range(1, num // 2 + 1):
    if num % i == 0:
        result.append(i)

result.append(num)  # include the number itself

print(result)


# TC = O(n/2) => O(n)
# SC = O(K) - k would be the total number of factors e.g. [1,2,5,10]
```

```python
# Method 3 (Optimal)

num = 10
result = []
for i in range(1, int(sqrt(num)) + 1):
    if num % i == 0:
        result.append(i)
        if i != num // i:  # To avoid adding the square root twice
            result.append(num // i)
result.sort()  # Optional: sort the factors in ascending order
print(result)
```

# Hashing Basics 
Store the frequency of every element in a Dictionary.

```python
lst = [1,2,3,4,5,2,4,6,3,3,3,3,3,3,3,]

dict = {}
for i in lst:
    if i in dict:
        dict[i] += 1
    else:
        dict[i] = 1

print(dict)

# TC = O(n)
# SC = O(n) - in worst case all elements are unique and we have to store them in the dictionary
```

```python
# Method 2 (Optimal)

dict = {}
for i in lst:
    dict[i] = dict.get(i, 0) + 1
print(dict) 

# TC = O(n)
# SC = O(n) - in worst case all elements are unique and we have to store them in the dictionary
```

```python
n = [5,3,2,2,1,5,5,7,5,10]
m = [10,111,1,9,5,67,2]

for num in n:
    if num in m:
        print(num, end = " ")
        
```

```python
for num in n:
    count = 0
    for x in m:
        if num == x:
            count += 1
    print(count, end = " ")


# TC = O(n*m)
# SC = O(1)

# Risk of TLE (Time Limit Exceeded) if the input size is large.
```

```python
# Method 2 (Optimal)
n = [5,3,2,2,1,5,5,7,5,10]
m = [10,111,1,9,5,67,2]

lst = [0] * 11
for num in n:
    lst[num] += 1
for num in m:
    if num < 1 or num > 10:      # 1 <= n[i] <= 10
        print(0)
    else:
        print(lst[num])


# TC = O(n + m)
# SC = O(1) - since the size of the list is fixed (11 in this case) regardless of the input size.

# | Query | Frequency |
# | ----- | --------- |
# | 10    | 1         |
# | 111   | 0         |
# | 1     | 1         |
# | 9     | 0         |
# | 5     | 4         |
# | 67    | 0         |
# | 2     | 2         |
```

```python
# Method 3 (Optimal) - Using Dictionary [Hash Map] and get() method

n = [5,3,2,2,1,5,5,7,5,10]
m = [10,111,1,9,5,67,2]

freq = {}

# Store frequency
for num in n:
    freq[num] = freq.get(num, 0) + 1

# Answer queries
for num in m:
    print(freq.get(num, 0), end=" ")
```

```python
# Character Hasing  

s = "azyxyyzaaaa"
q = ["d", "a", "y", "x"]     # output should be 0, 5, 3, 1
# constraint - only lowercase letters are present in the string  |  'a'<= s[i] <= 'z'

# Frequency array for 26 lowercase letters
hash_lst = [0] * 26

# Store frequencies
for ch in s:
    index = ord(ch) - ord('a')
    hash_lst[index] += 1

# Answer queries
for ch in q:
    index = ord(ch) - ord('a')
    print(hash_lst[index], end=" ")


# TC = O(n + q)
# SC = O(1)

```

Explanation
ASCII Mapping
ord(ch) - ord('a')

Converts characters into indices:

| Character | Index |
| --------- | ----- |
| a         | 0     |
| b         | 1     |
| c         | 2     |
| ...       | ...   |
| z         | 25    |



```python
# Method 2 (Optimal) - Using Dictionary [Hash Map] and get() method

s = "azyxyyzaaaa"
q = ["d", "a", "y", "x"]

freq = {}

for ch in s:
    freq[ch] = freq.get(ch, 0) + 1

for ch in q:
    print(freq.get(ch, 0), end=" ")
```

| List Hashing               | Dictionary Hashing      |
| -------------------------- | ----------------------- |
| Faster                     | Flexible                |
| Fixed size                 | Dynamic size            |
| Best for lowercase letters | Best for any characters |

# Recursion
Recursion is a programming technique where a function calls itself to solve a problem.

```python
def greet():
    print("Hello, World!")
    # greet()           # This will cause infinite recursion and eventually lead to a RecursionError: maximum recursion depth exceeded
greet()
```

### HEDA RECURSION

```python
# print hello 4 time using recursion

count = 0

def greet():
    global count
    
    if count == 4:
        return
    
    print("Hello")
    count += 1
    
    greet()

greet()
```

### TAIL RECURSION

```python
# print hello 4 time using recursion

count = 0

def greet():
    global count
    
    if count == 4:
        return
    
    count += 1
    greet()

    print("Hello")

greet()
```

### Recursion Using Parameters


```python
def func(x,n):
    if n == 0:
        return 
    print(x)
    func(x, n-1)

func(15,4)
```

```python
# Print 1 to n using recusion
# Head Recursion

def print_1_to_n(x,n):
    if x > n:
        return 
    
    print(x)
    print_1_to_n(x+1, n)


print_1_to_n(5, 15)
```

```python
# Tail Recursion / Backtracking Recursion

def print_1_to_n(x,n):
    if x > n:
        return 
    
    print_1_to_n(x+1, n)
    print(x)


print_1_to_n(5, 15)
```

```python
# Head Recursion

def printn(n):
    if n == 0:
        return 
    print(n)
    printn(n-1)
printn(5)
```

```python
# Tail Recursion / Backtracking Recursion

def printn(n):
    if n == 0:
        return 
    printn(n-1)
    print(n)
printn(5)
```

### Parameterized & Functional Recursion

```python
# Sum of n natural numbers using recursion

def sum_natural(n,count = 0):
    if n == 0:
        print(count)
        return
    
    count = count + n
    sum_natural(n-1, count)


sum_natural(10)
```

```python
# Sum of n natural numbers using recursion
# Functional Recursion

def sum_natural(n,count = 0):
    if n == 0:
        return count
    
    count = count + n
    return sum_natural(n-1, count)



print(sum_natural(10))
```

```python
# Functional Recursion
def func(n):
    if n == 1:
        return 1
    return n + func(n-1)

print(func(10))
```

```python
# Factorial 

def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)

print(factorial(5))
```

### Check if String is Palindrome 

```python
s = "nitin"

l = 0
r = len(s) - 1
ans = True

while l < r:
    if s[l] != s[r]:
        ans = False

    l += 1
    r -= 1

if ans == True:
    print("Palindrome")

else:
    print("NOt")

```

```python
s = "nsitisn"

def palin(s, left, right):
    if left >= right:
        # print("TRUE")
        return True
    
    if s[left] != s[right]:
        # print("False")
        return False
    
    palin(s, left + 1, right - 1)
    
palin(s, left = 0, right = len(s)-1)
```

## Selection Sort

```python
nums = [5,7,8,4,1,6,9,2]

# for Ascending Order
def selection_sort(nums):
    for i in range(len(nums)):
        min_index = i

        for j in range(i + 1, len(nums)):
            if nums[j] < nums[min_index]:
                min_index = j

        nums[i], nums[min_index] = nums[min_index], nums[i]

    return nums
 

print(selection_sort(nums))
```

```python
nums = [5,7,8,4,1,6,9,2]

# for Descending Order
def selection_sort(nums):
    for i in range(len(nums)):
        min_index = i

        for j in range(i + 1, len(nums)):
            if nums[j] > nums[min_index]:
                min_index = j

        nums[i], nums[min_index] = nums[min_index], nums[i]

    return nums


print(selection_sort(nums))
```

## Bubble Sort

```python
nums = [5,7,8,4,1,6,9,2]

def bubble_sort(nums):
    for i in range(len(nums)-2 , -1, -1 ):
        for j in range(0, i+1):
            if nums[j] > nums [j+1]:
                nums[j], nums[j+1] = nums[j+1], nums[j]
    return nums

bubble_sort(nums)
```

##### <b> Optimized Bubble Sort </b>

If no swapping happens in one pass, the list is already sorted, so we stop early.

| Case         | Complexity |
| ------------ | ---------- |
| Best Case    | O(n)       |
| Average Case | O(n²)      |
| Worst Case   | O(n²)      |

Best case becomes O(n) because of the optimization.

during a complete pass no swapping happens, it means:

nums[j] <= nums[j+1]

for every adjacent pair.

That means the entire array is already in sorted order.

```python
# Optimized Bubble Sort
nums = [5,7,8,4,1,6,9,2]

def bubbleSort(nums):
    
    for i in range (len(nums) - 2, -1, -1):
        is_swap = False
        
        for j in range(0, i+1):
            if nums[j] > nums[j+1]:
                nums[j], nums[j+1] = nums[j+1], nums[j]
                is_swap = True
                
        if is_swap == False:
            break
                
    return nums


print(bubble_sort(nums))
```
